1\. Считайте файл `pdf/rpd_tobd_22.pdf`. Извлеките из файла содержание дисциплины (раздел 5.1) в виде словаря, где ключ - это название темы, а значение - содержание темы. Выведите полученный словарь на экран. Выведите на экран количество тем курса.

```python
from PyPDF2 import PdfReader
import re

reader = PdfReader('/content/rpd_tobd_22.pdf')
pages = reader.pages
```

```python
text = ''
for page in pages:
  text+=page.extract_text()
  text = text.replace('\n', '')
text
```

```python
pattern = re.compile(r'5\.1\..*?5\.2\.')
text_51_52 = pattern.findall(text)[1][:-5]
text_51_52
```

```python
patt = re.compile(r'Тема\s\d\.')
split_51_52 = re.split(patt,text_51_52)[1:]
split_51_52
```

```python
res = {}
for i in split_51_52:
  key = re.split('\.', i)[0]
  value = re.split('\.', i)[1:]
  res[key] = '.'.join(value)
res
```

```python
len(res) # кол-во тем курса
```

4\. Считайте файл `docx/metodichka_azur_poluitog.docx`. Каждый абзац в тексте упражнения должен начинаться с символа "\t". Выведите полученную таблицу на экран.

```python
pip install python-docx
```

```python
from docx import Document
```

```python
doc = Document('/content/metodichka_azur_poluitog.docx')
```

```python
text = []
for paragraph in doc.paragraphs:
    text.append(paragraph.text)
text = '\t'.join(text).replace('\n', ' ')
text
```

```python
patt = re.compile(r'Упражнение\s\d')
ex = re.split(patt, text)[1:]
ex[0]
```

```python
imgs = []
patt_img = re.compile(r'Рис\.?\s?\d\.\d')
for i in ex:
  imgs.append(len(re.split(patt_img, i)) - 1)
imgs
```

```python
import numpy as np
df = pd.DataFrame(
    {
        'Номер упражнения': np.arange(1,8),
        'Текст упражнения': ex,
        'Кол-во рисунков в тексте': imgs
    }
)
df
```

```python
df['Текст упражнения'] = df['Текст упражнения'].apply(lambda x: x[2:])
df
```

5\. В файлах каталога `imgs` лежат 2 типа сканов: уведомления и пояснительные записки. Создайте `pd.DataFrame` с двумя столбцами: название файла и тип документа. Тип документа определите, воспользовавшись средством для распознавания текста и найдя ключевые слова.

```python
import pytesseract

imgs = ['б.jpg', 'в.jpeg', 'г.jpg', 'д.png', 'е.png', 'ж.png']

text = []
for i in imgs:
  img = Image.open(i)
  text.append(pytesseract.image_to_string(i, lang="rus"))

# text
```

```python
patt_note = re.compile(r'Пояснительная\sзаписка', re.IGNORECASE)
patt_uved = re.compile(r'Уведомление', re.IGNORECASE)
```

```python
note = 'Пояснительная записка'
uved = 'Уведомление'
type_doc = []
for i in text:
  if patt_note.findall(i):
    type_doc.append(note)
  elif patt_uved.findall(i):
    type_doc.append(uved)
type_doc
```

```python
df = pd.DataFrame({'Название файла': imgs, 'Тип документа': type_doc})
df
```
